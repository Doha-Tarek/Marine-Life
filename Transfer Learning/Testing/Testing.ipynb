```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.vgg16 import preprocess_input
from keras.models import load_model
from PIL import Image

# Load your saved VGG16 model
model_vgg = load_model('/content/drive/MyDrive/Underwater/Testing /VGG16_Training.h5')

# Create a generator for the test set
test_datagen_vgg = ImageDataGenerator(preprocessing_function=preprocess_input)
test_set_vgg = test_datagen_vgg.flow_from_directory(
    '/content/drive/MyDrive/Underwater/Testing /Test',
    target_size=(224, 224),
    batch_size=1,
    class_mode=None,
    shuffle=False
)

# Get predictions for the VGG model on the test set
predictions_vgg = model_vgg.predict_generator(test_set_vgg, steps=len(test_set_vgg.filenames))

# Map numerical predictions back to class labels
class_labels_vgg = list(test_set_vgg.class_indices.keys())

# Display predicted class labels for VGG
plt.figure(figsize=(15, 4))
for i in range(len(test_set_vgg.filenames)):
    img_path_vgg = test_set_vgg.filepaths[i]
    img_vgg = Image.open(img_path_vgg)
    img_vgg = img_vgg.resize((224, 224))  # Resize image to (224, 224)
    img_vgg = np.array(img_vgg)  # Convert image to array
    predicted_class_vgg = class_labels_vgg[np.argmax(predictions_vgg[i])]
    plt.subplot(1, len(test_set_vgg.filenames), i + 1)
    plt.imshow(img_vgg)
    plt.title(f"VGG Prediction: {predicted_class_vgg}")
    plt.axis('off')

plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Underwater/Testing /VGG_Predictions.png')
plt.show()
```
